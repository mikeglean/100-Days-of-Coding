```python
#mean of rows = axis = 1

array_hw_2 = np.array([[4,5],[7,10]])

np.mean(array_hw_2, axis =1)
```

```python
#adding data type as integer

np.mean(array_hw_2, axis = 0, dtype = int)
 
```

```python
#mean of columns = axis = 0

array_hw_2 = np.array([[4,5],[7,10]])

np.mean(array_hw_2, axis =0)
```
